```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA 
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier ,RandomForestClassifier ,GradientBoostingClassifier 
```

```python
data = pd.read_csv(r"C:\Users\Krutarth\Desktop\Datasets\greyatom_forest\data.csv")
data_scaled = pd.read_csv(r"C:\Users\Krutarth\Desktop\Datasets\greyatom_forest\data_scaled.csv")
data_scaled
```

```python
#outliers-- no results
#feature selection-- pca, rfe, chi2, anova--anova at 5 features is the best
#different models
#feature engineering
#different scalers
```

```python
selector = SelectKBest(k=4)
data_feature = pd.DataFrame(selector.fit_transform(data_scaled.drop(columns=['Cover_Type']),data.Cover_Type))
data_feature
```

```python
data_pred = pd.merge(data_feature, data.drop(columns=data_scaled.columns),on = data_scaled.index)
data_pred.drop(columns=['key_0'], inplace=True)
data_pred
```

```python
X = data_pred
y = data.Cover_Type
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
algorithms = [LogisticRegression(), DecisionTreeClassifier(), GradientBoostingClassifier(), RandomForestClassifier(), GradientBoostingClassifier() ]
```

```python
scores = []
for algorithm in algorithms:
    model = algorithm
    model = model.fit(X_train, y_train)

    score = model.score(X_test, y_test)
    scores.append([str(algorithm), score])
scores
```

```python
log_reg = RandomForestClassifier()
log_reg = log_reg.fit(X_train, y_train)

score = log_reg.score(X_test, y_test)
score
```

```python
test = pd.read_csv(r"C:\Users\Krutarth\Desktop\Datasets\greyatom_forest\test.csv")
test_scaled = pd.read_csv(r"C:\Users\Krutarth\Desktop\Datasets\greyatom_forest\test_scaled.csv")
Id = test.Id
test.drop(columns=['Id'], inplace=True)
```

```python
test_scaled
```

```python
test['Hillshade'] = test.Hillshade_9am + 2*test.Hillshade_Noon + test.Hillshade_3pm
test.drop(columns=['Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm'], inplace=True)
```

```python
test['Distance_To_Hydrology'] = np.sqrt(test.Horizontal_Distance_To_Hydrology**2 + test.Vertical_Distance_To_Hydrology**2)
test.drop(columns=['Horizontal_Distance_To_Hydrology','Vertical_Distance_To_Hydrology'], inplace=True)
```

```python
test_feature = pd.DataFrame(selector.transform(test_scaled))
```

```python
test_feature
```

```python
test_pred = pd.merge(test_feature, test.drop(columns=data_scaled.drop(columns=['Cover_Type']).columns),on = test_scaled.index)
test_pred.drop(columns=['key_0'], inplace=True)
```

```python
test_pred
```

```python
predictions = log_reg.predict(test_pred)
predictions
```
