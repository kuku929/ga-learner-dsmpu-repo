```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X_col = x.columns
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(x))
X.columns = X_col
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
grid_search = GridSearchCV(estimator=classifier, param_grid = parameter_grid,n_jobs=20)
```
